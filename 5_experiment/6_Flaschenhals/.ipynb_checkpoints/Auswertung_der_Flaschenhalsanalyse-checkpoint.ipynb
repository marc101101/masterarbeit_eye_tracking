# Auswertung der Flaschenhalsanalyse

Das System wurde so aufgebaut, dass sich alle Webcams direkt nebeneinander befanden und zur selben Zeit ein Gesicht gleichzeitig erfassen konnten. Dann wurden über ca. 90 Minuten die Zeit-Logausgaben gesammelt und ausgewertet, wobei 67.393 Datensätze erzeugt und anschließend deren zeitlicher Abstand zueinander bestimmt wurden (s. Anhang). Die nachfolgende Abbildung zeigt alle Verarbeitungsschritte und die jeweiligen Dauer der einzelnen Schritte.

```python
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt

with open("logOLD.txt") as f:
    lines = f.readlines()
    print(len(lines))
```

```python
elements = [["Eingangsbild der Webcam", 0, 0, 0],
    ["Gesichtserkennung", 0, 0, 0],
    ["Gesichts-Landmark-Erkennung", 0, 0, 0],
    ["Schätzung des Blicks", 0, 0, 0],
    ["Schätzung der Kopfhaltung", 0, 0, 0],
    ["Log-Ausgabe an die Konsole", 0, 0, 0],
    ["Weiterleitung an Websocket", 0, 0, 0],
    ["Annahme der Blickdaten",  0, 0, 0],
    ["Validierung", 0, 0, 0],
    ["Persistierung", 0, 0, 0],
    ["Transformation",  0, 0, 0],
    ["Schnittpunkt-Berechnung", 0, 0, 0],
    ["Cross Hair Abweichung",  0, 0, 0],
    ["Weiterleitung der Daten an Websocket",  0, 0, 0],
    ["Annahme der transformierten Daten",  0, 0, 0],
    ["Visualisierung der Daten",  0, 0, 0]]
```

```python
counter = 1
counter_max = 67391 

while(counter <= counter_max):
    current_time = lines[counter][0:23]
    current_old = lines[counter-1][0:23]
    current_station = lines[counter][49:90]
    
    current = datetime.strptime(current_time, '%Y-%m-%d %H:%M:%S.%f')
    old = datetime.strptime(current_old, '%Y-%m-%d %H:%M:%S.%f')
    diff = (current-old).total_seconds()
    
    if(diff > 0):
        element_counter = 0
        for element in elements:
            if(element[0] == current_station[:-1]):
                element[1] = diff + element[1]
                element[2] = element[2] + 1
            
            element_counter = element_counter + 1
    
    counter = counter +1
    
for element in elements:
    element[3] = element[1] / element[2]
```

```python
plt.rcParams["figure.figsize"] = (17, 5) # (w, h)

df = pd.DataFrame(elements)

df.plot(kind='bar', x=0, y=3)
```
